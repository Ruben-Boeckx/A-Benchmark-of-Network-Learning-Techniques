```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
G = nx.Graph()

G.add_node('C1', node_type='client')
G.add_node('C2', node_type='client')
G.add_node('C3', node_type='client')
G.add_node('T1', node_type='transaction')
G.add_node('T2', node_type='transaction')
G.add_node('T3', node_type='transaction')
G.add_node('T4', node_type='transaction')
G.add_node('T5', node_type='transaction')
G.add_node('T6', node_type='transaction')
G.add_node('T7', node_type='transaction')
G.add_node('M1', node_type='merchant')
G.add_node('M2', node_type='merchant')
G.add_node('M3', node_type='merchant')
G.add_node('M4', node_type='merchant')
G.add_node('M5', node_type='merchant')

G.add_edges_from([
    ('C1', 'T1'), ('T1', 'M1'),
    ('C1', 'T2'), ('T2', 'M2'),
    ('C1', 'T3'), ('T3', 'M2'),
    ('C2', 'T4'), ('T4', 'M3'),
    ('C2', 'T5'), ('T5', 'M5'),
    ('C3', 'T6'), ('T6', 'M4'),
    ('C3', 'T7'), ('T7', 'M5')
])

pos = {
    'C1': (0, 1.5), 'C2': (0, 0.75), 'C3': (0, 0),
    'T1': (1, 1.8), 'T2': (1, 1.5), 'T3': (1, 1.2),
    'T4': (1, 0.9), 'T5': (1, 0.6),
    'T6': (1, 0.3), 'T7': (1, 0),
    'M1': (2, 1.8), 'M2': (2, 1.35), 'M3': (2, 0.9),
    'M4': (2, 0.45), 'M5': (2, 0)
}
```

```python
plt.figure(figsize=(8, 6))

nx.draw_networkx_nodes(G, pos, nodelist=[n for n, d in G.nodes(data=True) if d['node_type'] == 'client'],
                       node_color='lightblue', node_size=1500, node_shape='o')
nx.draw_networkx_nodes(G, pos, nodelist=[n for n, d in G.nodes(data=True) if d['node_type'] == 'transaction'],
                       node_color='lightgreen', node_size=1000, node_shape='s')
nx.draw_networkx_nodes(G, pos, nodelist=[n for n, d in G.nodes(data=True) if d['node_type'] == 'merchant'],
                       node_color='salmon', node_size=1500, node_shape='h')

nx.draw_networkx_edges(G, pos, edge_color='gray', width=1)

labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels, font_size=8, font_weight='bold')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Client',
           markerfacecolor='lightblue', markersize=10),
    Line2D([0], [0], marker='s', color='w', label='Transaction',
           markerfacecolor='lightgreen', markersize=10),
    Line2D([0], [0], marker='h', color='w', label='Merchant',
           markerfacecolor='salmon', markersize=10)
]

plt.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.25, 1))

plt.axis('off')
plt.tight_layout()
plt.show()
```
